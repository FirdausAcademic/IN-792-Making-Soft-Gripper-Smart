```python
# Set up the serial connection (adjust 'COM7' to your Arduino's port)
# ser = serial.Serial('COM7', 250000, timeout=1)
# ser.flush()

test_file = 'gripper_data_cup10_torq_1.csv'
#sensor_file = 'sensor_data_3DprintedTetrahedron10_torq_1.csv'

input_torque = 1
duration = 5  # seconds
iterations = 10 # number of iterations

all_test_data = []
all_sensor_data = []

# Data lists
time_values = []
desired_torque = []
output_value = []
encoder_values = []
# sensor1_values = []
# sensor2_values = []
# sensor3_values = []

# Start time
start_time = time.time()

odrv0.axis0.encoder.set_linear_count(0)

try:
    for _ in range(iterations):
        while True:
            for i in range(200):
                odrv0.axis0.controller.input_torque = input_torque
                torque = odrv0.axis0.motor.current_control.Iq_measured * odrv0.axis0.motor.config.torque_constant
                encoder_value = odrv0.axis0.encoder.pos_estimate

                # Store time and chirp signal values
                t = time.time() - start_time
                time_values.append(t)
                desired_torque.append(input_torque)
                output_value.append(torque)
                encoder_values.append(encoder_value)

                # line = ser.readline().decode('latin-1').rstrip()
                # values = line.split(',')
                # print(values)

                # if len(values) == 3:  # We expect three sensor values
                #     sensor1_values.append(values[0])
                #     sensor2_values.append(values[1])
                #     sensor3_values.append(values[2])
                #     print(f"Logged data: {values}")
                # i = i + 1
                # Add delay between iterations
            break

        odrv0.axis0.controller.input_torque = 0
        time.sleep(1)  # Add delay between iterations

    # Combine all test data into a single list
    for i in range(len(time_values)):
        test_row = [
            desired_torque[i],
            output_value[i],
            encoder_values[i]
        ]
        all_test_data.append(test_row)

    # Combine all sensor data into a single list
    # for i in range(len(time_values)):
    #     sensor_row = [
    #         sensor1_values[i] if i < len(sensor1_values) else None,
    #         sensor2_values[i] if i < len(sensor2_values) else None,
    #         sensor3_values[i] if i < len(sensor3_values) else None
    #     ]
    #     all_sensor_data.append(sensor_row)

except KeyboardInterrupt:
    print("Interrupted by user")

# Create DataFrames with the combined data
test_columns = ['Desired Torque', 'Output Torque', 'Encoder']
# sensor_columns = ['Sensor1', 'Sensor2', 'Sensor3']
test_df = pd.DataFrame(all_test_data, columns=test_columns)
# sensor_df = pd.DataFrame(all_sensor_data, columns=sensor_columns)

# Save the DataFrames to CSV files
try:
    test_df.to_csv(test_file, index=False)
    print(f"Test data saved to {test_file}")
except PermissionError:
    print(f"Permission denied: unable to save to '{test_file}'. Please close the file if it is open or save it with a different name.")
    backup_test_file = 'test_backup.csv'
    test_df.to_csv(backup_test_file, index=False)
    print(f"Test data saved to {backup_test_file}")

# try:
#     sensor_df.to_csv(sensor_file, index=False)
#     print(f"Sensor data saved to {sensor_file}")
# except PermissionError:
#     print(f"Permission denied: unable to save to '{sensor_file}'. Please close the file if it is open or save it with a different name.")
#     backup_sensor_file = 'sensor_data_backup.csv'
#     sensor_df.to_csv(backup_sensor_file, index=False)
#     print(f"Sensor data saved to {backup_sensor_file}")

# Get current working directory and create file paths
cwd = os.getcwd()
test_file_path = os.path.join(cwd, test_file)
# sensor_file_path = os.path.join(cwd, sensor_file)
print(f"Test data file is located at {test_file_path}")
# print(f"Sensor data file is located at {sensor_file_path}")

# ser.close()
```
