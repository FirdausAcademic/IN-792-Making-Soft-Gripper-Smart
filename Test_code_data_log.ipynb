```python
import serial
import pandas as pd
from datetime import datetime
import csv
import time

```

```python
ser = serial.Serial('COM7', 250000, timeout=1)
ser.flush()

data = []


try:
    while True:
        if ser.in_waiting > 0:
            line = ser.readline().decode('latin-1').rstrip()
            values = line.split(',')
            if len(values) == 3:  # We expect three sensor values
                data.append(values)
                print(f"Logged data: {values}")

except KeyboardInterrupt:
    # When the user interrupts (Ctrl+C), save the data to an Excel file
    df = pd.DataFrame(data, columns=['Sensor1', 'Sensor2', 'Sensor3'])
    try:
        df.to_excel('arduino_data.xlsx', index=False)
        print("Data saved to arduino_data.xlsx")
    except PermissionError:
        print("Permission denied: unable to save to 'arduino_data.xlsx'. Please close the file if it is open or save it with a different name.")
        # Save with a different name
        df.to_excel('arduino_data_backup.xlsx', index=False)
        print("Data saved to arduino_data_backup.xlsx")
finally:
    ser.close()
```

```python


# Set the serial port and baud rate
serial_port = 'COM7'  # Change this to your Arduino's serial port
baud_rate = 250000

# Open the serial port
ser = serial.Serial(serial_port, baud_rate, timeout=1)
ser.flush()

data = []

try:
    while True:
        if ser.in_waiting > 0:
            line = ser.readline().decode('latin-1').rstrip()
            values = line.split(',')
            if len(values) == 3:  # We expect three sensor values
                data.append(values)
                print(f"Logged data: {values}")

except KeyboardInterrupt:
    # When the user interrupts (Ctrl+C), save the data to a CSV file
    try:
        with open('sensor_data_cup10_torq_1.csv', mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Sensor1', 'Sensor2', 'Sensor3'])  # Write the header
            writer.writerows(data)  # Write the data
        print("Data saved to arduino_data.csv")
    except PermissionError:
        print("Permission denied: unable to save to 'arduino_data.csv'. Please close the file if it is open or save it with a different name.")
        # Save with a different name
        with open('arduino_data_backup.csv', mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['Sensor1', 'Sensor2', 'Sensor3'])  # Write the header
            writer.writerows(data)  # Write the data
        print("Data saved to arduino_data_backup.csv")
finally:
    ser.close()
```
